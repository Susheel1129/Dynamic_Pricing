```python
# 03_inventory_check.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_DIR = "../data"

sales = pd.read_csv(f"{DATA_DIR}/sales_sim.csv")
inventory = pd.read_csv(f"{DATA_DIR}/inventory.csv")

inventory['shortfall'] = (sales['units_sold'] - inventory['inventory']).clip(lower=0)

# Summary
print(inventory.head())

# Visualize shortfall
plt.figure(figsize=(10,6))
sns.histplot(inventory['shortfall'], bins=20, kde=False)
plt.xlabel("Shortfall Units")
plt.ylabel("Number of SKUs")
plt.title("Inventory Shortfall Distribution")
plt.show()
```
